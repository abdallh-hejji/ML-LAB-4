```python
import pandas as pd

# Load the data and fix the data types (from the lab's earlier steps)
df = pd.read_csv("Chocolate_Sales.csv")
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df['Amount'] = df['Amount'].replace(r'[\$,]', '', regex=True)
df['Amount'] = pd.to_numeric(df['Amount'])

# Now run Task 1
print("Missing Values:\n", df.isna().sum())
print("\nData Types:\n", df.dtypes)
print("\nStatistical Summary:\n", df.describe())
```

```python
#Task 1
# Check for missing values in each column
print("Missing Values:\n", df.isna().sum())

# Check data types (to ensure Date is datetime and Amount is numeric)
print("\nData Types:\n", df.dtypes)

# Get statistical summary to spot potential anomalies (e.g., negative amounts or crazy max values)
print("\nStatistical Summary:\n", df.describe())
```

```python
#Task 2
# Creating a copy to keep the original intact
df_clean = df.copy()

# Applying Median Imputation to the 'Amount' column (Updated for Pandas 3.0)
median_value = df_clean['Amount'].median()
df_clean['Amount'] = df_clean['Amount'].fillna(median_value)

print("Missing values after imputation: ", df_clean['Amount'].isna().sum())
```

```python
#Task 3 

# Calculate Q1, Q3, and IQR for the 'Amount' column
Q1 = df_clean['Amount'].quantile(0.25)
Q3 = df_clean['Amount'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Handle outliers by filtering them out (Removing)
df_no_outliers = df_clean[(df_clean['Amount'] >= lower_bound) & (df_clean['Amount'] <= upper_bound)]

print("Rows before removing outliers:", df_clean.shape[0])
print("Rows after removing outliers:", df_no_outliers.shape[0])
```

```python
#Task 4

from sklearn.preprocessing import MinMaxScaler, StandardScaler

# 1. Min-Max Normalization
minmax_scaler = MinMaxScaler()
df_minmax = df_no_outliers[['Amount', 'Boxes Shipped']].copy()
df_minmax[['Amount', 'Boxes Shipped']] = minmax_scaler.fit_transform(df_minmax)

print("Min-Max Normalized Data:\n", df_minmax.head())

# 2. Z-score Normalization (Standardization)
standard_scaler = StandardScaler()
df_zscore = df_no_outliers[['Amount', 'Boxes Shipped']].copy()
df_zscore[['Amount', 'Boxes Shipped']] = standard_scaler.fit_transform(df_zscore)

print("\nZ-score Standardized Data:\n", df_zscore.head())
```

```python
#Task 5
from sklearn.decomposition import PCA

# Apply PCA keeping 2 components
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df_zscore[['Amount', 'Boxes Shipped']])

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio for PC1 and PC2:", explained_variance)
```
